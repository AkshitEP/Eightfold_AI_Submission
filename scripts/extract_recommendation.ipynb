```python
def llm_invoke(system_prompt,msg):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": system_prompt,
            },
            {
                "role": "user",
                "content": msg,
            }
        ],
        model="llama-3.1-70b-versatile",
        temperature=0
    )

    return (chat_completion.choices[0].message.content)
    
```
